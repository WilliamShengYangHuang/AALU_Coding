```python
import random
import matplotlib.pyplot as plt
import csv
```

```python
# Generate a list of 100 random coordinates
coordinates = [(random.uniform(1, 100), random.uniform(1, 100)) for _ in range(100)]

# Sort the coordinates based on the values of x and y
# If two coordinates have the same x value, they are sorted based on their y value.
sorted_coordinates = sorted(coordinates, key=lambda coord: (coord[0], coord[1]))

# Arrange the sorted coordinates into a 10x10 matrix
# This is achieved by slicing the sorted list every 10 elements to create rows for the matrix.
matrix = [sorted_coordinates[i:i + 10] for i in range(0, 100, 10)]

matrix # Display the matrix result
```

```python
# Create a scatter plot
x_values = [coord[0] for coord in coordinates]  # Extract x-values
y_values = [coord[1] for coord in coordinates]  # Extract y-values

plt.scatter(x_values, y_values)  # Plot x and y values

# Annotate each point with its index
for i, (x, y) in enumerate(coordinates):
    plt.text(x, y, str(i), fontsize=8)

plt.title('Scatter Plot of Random Coordinates with Indices')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()  # Display the plot
```

```python
# Save the matrix to a CSV file
with open('coordinates_matrix.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    for row in matrix:
        writer.writerow(row)
```
